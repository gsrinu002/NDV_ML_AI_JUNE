```python
# app.py
import streamlit as st
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset (example: Iris)
@st.cache_data
def load_data():
    return sns.load_dataset('iris')

df = load_data()

# Title
st.title("ğŸŒ¼ Iris Dataset Dashboard")

# Sidebar filters
st.sidebar.header("Filter Options")
species = st.sidebar.multiselect("Select Species", df['species'].unique(), default=df['species'].unique())

# Apply filters
filtered_df = df[df['species'].isin(species)]

# Show filtered data
st.subheader("ğŸ” Filtered Data")
st.dataframe(filtered_df)

# Plot 1: Pairplot
st.subheader("ğŸ“ˆ Pairplot")
fig1 = sns.pairplot(filtered_df, hue='species')
st.pyplot(fig1)

# Plot 2: Correlation Heatmap
st.subheader("ğŸ“Š Correlation Heatmap")
fig2, ax = plt.subplots()
sns.heatmap(filtered_df.drop('species', axis=1).corr(), annot=True, cmap="coolwarm", ax=ax)
st.pyplot(fig2)

# Metrics
st.subheader("ğŸ“Œ Summary Statistics")
st.write(filtered_df.describe())
```
