````python
# conditions
# conditions are checking whether statement is true or false
# if  : if condition checks only true statements.
# if-else : if-else condition checks true and false blocks.
# elif (else if) : checks more and more true statements
# nested if : if inside if (condition inside condition)

````

````python
# if  : if condition checks only true statements.

name ="Python"
if name=="Python":
  print("matched")
  print("both names are equal")
````

````python
# if-else : if-else condition checks true and false blocks.
name ="Python"
if name=="python":
  print("matched")
  print("both names are equal")
else:
  print("not matched")
  print("both names are not equal")
````

````python
# elif (else if) : checks more and more true statements
speed=10

if speed<30:
  print("low speed")
elif speed>30 and speed<60:
  print("avg speed")
elif speed>60 and speed<100:
  print("good speed")
else:
  print("high speed")
````

````python
# nested if : if inside if (condition inside condition)
qualified=True
doingJob=True

if qualified==True:
  print("Qualified")
  if doingJob==True:
    print("good carryon")
  else:
    print("try for good job")
else:
  print("not qualified")
````

````python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('netflix_titles.csv')
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Handle missing values
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Not Specified', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna(df['rating'].mode()[0], inplace=True)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract year, month, and weekday from date_added
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df['weekday_added'] = df['date_added'].dt.day_name()

# Split 'listed_in' (genres) into lists
df['listed_in_list'] = df['listed_in'].apply(lambda x: [i.strip() for i in x.split(',')] if pd.notnull(x) else [])

# Flag for multiple genres
df['multi_genre'] = df['listed_in_list'].apply(lambda x: len(x) > 1)

# Create content length feature (dummy: e.g., for Movies - 1, TV Shows - 10)
df['content_length'] = np.where(df['type'] == 'Movie', 1, 10)

# Handle inconsistent country names (example)
df['country'] = df['country'].replace({'United States of America': 'United States'})

# Outlier check for release years
df = df[(df['release_year'] >= 1920) & (df['release_year'] <= 2025)]

# Aggregation example: Number of titles by country
country_counts = df['country'].value_counts().head(10)
print("\nTop 10 Countries by Content:\n", country_counts)

# Final dataset preview
print("\nCleaned Netflix Dataset:")
print(df.head())
````
