````python
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount("/content/drive")
# Load data
df = df = pd.read_csv('/content/drive/MyDrive/Data/netflix_titles.csv')

# 1ï¸âƒ£ Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# 2ï¸âƒ£ Remove duplicates
df.drop_duplicates(inplace=True)

# 3ï¸âƒ£ Handle missing values
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Not Specified', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna(df['rating'].mode()[0], inplace=True)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# 4ï¸âƒ£ Feature Engineering
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df['weekday_added'] = df['date_added'].dt.day_name()

df['multi_genre_flag'] = df['listed_in'].apply(lambda x: ',' in str(x))
df['num_genres'] = df['listed_in'].apply(lambda x: len(str(x).split(',')) if pd.notnull(x) else 0)





# 5ï¸âƒ£ Fix inconsistent entries (e.g., country names)
df['country'] = df['country'].replace({
    'United States of America': 'United States',
    'United Kingdom of Great Britain': 'United Kingdom'
})

# 6ï¸âƒ£ Filter out extreme outliers
df = df[df['release_year'].between(1920, 2025)]

# 7ï¸âƒ£ Final overview
print("Final Shape:", df.shape)
print("\nSample Data:\n", df.head())
print("\nMissing Values:\n", df.isnull().sum())
````

````python
# Visualization

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the preprocessed dataset
df = pd.read_csv("netflix_cleaned.csv")

# Set a visual style
sns.set(style="whitegrid")

# Plot 1: Distribution of content types (Movies vs TV Shows)
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='type', palette='Set2')
plt.title('Distribution of Content Type on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Plot 2: Top 10 countries with most content
plt.figure(figsize=(10, 6))
top_countries = df['country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette='mako')
plt.title('Top 10 Countries with Most Content on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.tight_layout()
plt.show()

# Plot 3: Number of titles added over time
plt.figure(figsize=(12, 6))
df['year_added'] = pd.to_datetime(df['date_added']).dt.year
yearly_counts = df['year_added'].value_counts().sort_index()
sns.lineplot(x=yearly_counts.index, y=yearly_counts.values, marker='o')
plt.title('Number of Titles Added Over Time')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.tight_layout()
plt.show()

# Plot 4: Most common genres (from 'listed_in')
from collections import Counter

all_genres = df['listed_in'].dropna().str.split(', ')
flat_genres = [genre for sublist in all_genres for genre in sublist]
genre_counts = Counter(flat_genres).most_common(10)

genres, counts = zip(*genre_counts)
plt.figure(figsize=(10, 6))
sns.barplot(x=counts, y=genres, palette='coolwarm')
plt.title('Top 10 Most Common Genres on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()

# Plot 5: Duration distribution of movies (in minutes)
plt.figure(figsize=(10, 6))
movie_durations = df[(df['type'] == 'Movie') & (df['duration_unit'] == 'min')]['duration_int']
sns.histplot(movie_durations, bins=30, kde=True, color='purple')
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()
````
